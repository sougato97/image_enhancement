```python
# Import the required libraries
import cv2
import matplotlib.pyplot as plt

# Load the image and convert it to grayscale
img = cv2.imread('/home/sougato97/Thesis/datasets/LOLdataset/eval15/high/111.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create an ORB object with the desired number of keypoints and other parameters
orb = cv2.ORB_create(nfeatures=5000)

# Use the detect_and_extract method of the ORB object to find the keypoints and descriptors of the image
kp, des = orb.detectAndCompute(gray, None)

# Use the plot_matches function from skimage.feature to display the keypoints on the image
plt.figure(figsize=(10, 10))
plt.imshow(cv2.drawKeypoints(gray, kp, None, color=(0, 255, 0)))
plt.title('ORB features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Define a range of values for x from 0 to 10 with a step of 0.1
x = [i * 0.1 for i in range (0, 101)]

# Apply the sqrt () function to each x value to get the y values
y = [math.sqrt (x_i) for x_i in x]

# Plot the graph of square root using plt.plot ()
plt.plot (x, y)

# Add labels and title
plt.xlabel ('Signal')
plt.ylabel ('Shot Noise  (signal)')
plt.title ('Shot Noise v/s Signal')

# Show the plot
plt.show ()
```
